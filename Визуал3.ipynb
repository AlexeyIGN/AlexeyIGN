```python
import pandas as pd
import datetime
from matplotlib import pyplot as plt
import seaborn as sns
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
! pip install plotly
%matplotlib inline

from google.colab import files
 
uploaded = files.upload()
import io
df = pd.read_excel(io.BytesIO(uploaded['b9342fd9-2a47-4f5c-b26c-ddde8fb453bd.xls']))

df.purchase_dt = df.purchase_dt.where(df.purchase_dt.notnull(), other=-1)
df = df.loc[df['purchase_dt'] != -1].reset_index()
df= df.drop(['index'], axis = 1)
df['purchase_dt'] = pd.to_datetime(df['purchase_dt'],errors='coerce')
df['UTime'] = (df['purchase_dt'] - df['reg_dttm']).dt.days
def f1(df):
  return df.sort_values(by = ['UTime']).head(1)
d1 = df.groupby(['client_id']).apply(f1)
d1.sort_values(by = ['UTime']).head()

d1.UTime = d1.UTime.where(d1.UTime != -1, other = 0)
d1.sort_values(by = ['UTime']).head()

def f2(d1):
  return d1['purchase_amt'].mean()
def f3(d1):
  return d1['purchase_amt'].sum()

d2 = d1.groupby(['UTime']).apply(f2).to_frame(name='AvgPur').reset_index()
d2.head()

d3 = d1.groupby(['UTime']).apply(f3).to_frame(name='SumPur').reset_index()
d3.head()

fig1 = go.Figure()
fig1.add_trace(go.Bar(x = d2['UTime'], y=d2['AvgPur']))
fig1.update_layout( title="Как скорость утилизации влияет на размер первой покупки", yaxis_title="Средний Размер первой покупки, руб",
                 xaxis_title="Скорость утилизации, дни")
fig1.show()

fig = go.Figure()
fig.add_trace(go.Bar(x = d3['UTime'], y=d3['SumPur']))
fig.update_layout( title="Как скорость утилизации влияет на размер первой покупки", yaxis_title="Сумма первых покупок, руб",
                 xaxis_title="Скорость утилизации, дни")
fig.show()
```
